### Fire Fighters versus Police Officers

- Using the Salaries.csv file calculate:
- What is the ratio between people in the fire department over people in the police department
- What is the mean salary of the police department? Use the "BaseSalary" column.
- What is the mean salary of the fire department? Use the "BaseSalary" column.

```python
salaries_df = pd.read_csv('Salaries.csv', index_col=0)
job_title = salaries_df.iloc[2]

# Define a function to return all rows containing the job title 'POLICE'

def find_police(x):
    return "POLICE" in x

salaries_df['isPolice'] = salaries_df['JobTitle'].apply(find_police)
salaries_df[['JobTitle','isPolice']]

police_dep = salaries_df['isPolice'].sum()

# Define a function to return all rows containing the job title 'POLICE'

def find_fire(x):
    return "FIRE" in x

salaries_df['isFire'] = salaries_df['JobTitle'].apply(find_fire)
salaries_df[['JobTitle','isFire']]

fire_dep = salaries_df['isFire'].sum()

# Ratio of people in fire department to police department 

print(fire_dep / police_dep)
```

```python
# Calculate mean salary of police department

police_df = salaries_df[salaries_df['isPolice']]
police_mean_salary = police_df['BasePay'].sum() / police_dep

fire_df = salaries_df[salaries_df['isFire']]
fire_mean_salary = fire_df['BasePay'].sum() / fire_dep
print("The average salary for the police department is {0:.2f} and the average salary for the fire department is {1:.2f}".format(police_mean_salary, fire_mean_salary))
```

### DataFrame Manipulation

- How many unique jobs are there? Look at the unique method in pandas
- Count how many people have the first name "John"
- Find the surname of everyone whose first name is more than 6 characters long

```python
unique_jobs = salaries_df['JobTitle'].nunique()
print(unique_jobs)

def find_john(x):
    return 'JOHN' in x

salaries_df['isJohn'] = salaries_df['EmployeeName'].apply(find_john)
salaries_df['isJohn'].sum()

def surnames(x):
    return x.split()[1]

salaries_df['Surname'] = salaries_df['EmployeeName'].apply(surnames)

def surname_letters(x):
    return len(x) > 6

salaries_df['Surname'] = salaries_df['Surname'].apply(surname_letters)
salaries_df[salaries_df['Surname']]
```
